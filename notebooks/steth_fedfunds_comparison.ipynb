# Average APY: stETH vs Fed Funds Rate
Calculates the average percentage over the period May 2024 – May 2025.

```python
import pandas as pd
```

```python
steth = pd.read_csv('../data/steth_apy.csv', parse_dates=['DATE'])
fedfunds = pd.read_csv('../data/fedfunds.csv', parse_dates=['observation_date'])

# Standardise column names for merging
steth = steth.rename(columns={'DATE': 'date'})
fedfunds = fedfunds.rename(columns={'observation_date': 'date'})

df = pd.merge(steth, fedfunds, on='date')
df
```

```python
avg_steth    = round(df['stETH_APY'].mean(), 2)
avg_fedfunds = round(df['FEDFUNDS'].mean(), 2)

print(f"Average stETH APY  : {avg_steth:.2f}%")
print(f"Average Fed Funds  : {avg_fedfunds:.2f}%")
```
